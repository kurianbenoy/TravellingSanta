```python
def write_tsp(cities, filename, name='traveling-santa-2018-prime-paths'):
    with open(filename, 'w') as f:
        f.write('NAME : %s\n' % name)
        f.write('COMMENT : %s\n' % name)
        f.write('TYPE : TSP\n')
        f.write('DIMENSION : %d\n' % len(cities))
        f.write('EDGE_WEIGHT_TYPE : EUC_2D\n')
        f.write('NODE_COORD_SECTION\n')
        for row in cities.itertuples():
            f.write('%d %.11f %.11f\n' % (row.Index+1, row.X, row.Y))
        f.write('EOF\n')

write_tsp(cities1k, 'cities1k.tsp')
```

```python
class Tour(list):  # list with trailing 0
    penalized = ~cities.index.isin(sympy.primerange(0, len(cities)))

    @staticmethod
    def from_file(filename):  # from linkern's output or csv
        seq = [int(x) for x in open(filename).read().split()[1:]]
        return Tour(seq if seq[-1] == 0 else (seq + [0]))

    def score(self):
        df = cities.reindex(self)
        dist = np.hypot(df.X.diff(-1), df.Y.diff(-1))
        penalty = 0.1 * dist[9::10] * self.penalized[self[9::10]]
        return dist.sum() + penalty.sum()

    def to_csv(self, filename):
        pd.DataFrame({'Path': self}).to_csv(filename, index=False)
```

```python
tour = Tour.from_file('linkern.tour')
tour.to_csv('submission.csv')
tour.score()
```

```python
def plot_tour(tour, cmap=mpl.cm.gist_rainbow):
    fig, ax = plt.subplots(figsize=(25, 25/1653*1283))
    n = len(tour)

    for i in range(200):
        ind = tour[n//200*i:min(n, n//200*(i+1)+1)]
        plt.plot(cities.X[ind], cities.Y[ind], color=cmap(i/200.0), linewidth=1)

    ax.autoscale(tight=True)
    mpl.colorbar.ColorbarBase(ax=fig.add_axes([0.125, 0.075, 0.775, 0.01]),
                              norm=mpl.colors.Normalize(vmin=0, vmax=n),
                              cmap=cmap, orientation='horizontal')

plot_tour(tour)
```
